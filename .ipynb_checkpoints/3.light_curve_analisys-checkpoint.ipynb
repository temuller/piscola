```python
def decompose(algorithm, features, phases, weights=None, n_components=3):
    
    kwargs = {}
    if (weights is not None) and ('sklearn' not in str(algorithm)) and ('umap' not in str(algorithm)):
        kwargs = {'weights':weights}
        
    if ('sklearn' in str(algorithm)) or ('wpca' in str(algorithm)) or ('umap' in str(algorithm)):
        reducer = algorithm(n_components=n_components)
        transformed_features = reducer.fit_transform(features, **kwargs) # fits AND transforms at the same time
        try:
            components_ = reducer.components_
        except:
            pass
    else:
        reducer = algorithm(feature_values, n_components=n_components, V=weights)
        reducer.SolveNMF()
        transformed_features = reducer.W
        components_ = reducer.H

    color_palette = [plt.get_cmap('Dark2')(i) for i in np.arange(8)] + [plt.get_cmap('Set1')(i) for i in np.arange(8)]
    color_palette = color_palette[::3]

    if ('umap' not in  str(algorithm)) and ('TSNE' not in  str(algorithm)):
        fig, ax = plt.subplots(figsize=(8, 6))
        for i in np.arange(n_components):
            variance = transformed_features[:,i].var()/np.sum(transformed_features.var(axis=0))
            component_array = components_[i,:]
            ax.plot(phases, component_array, 'o-', 
                    label = f'NMF comp. {i+1} ({int(np.round(100*variance, 0))}%)', color=color_palette[i])

        ax.set_ylabel('Component value')
        ax.set_xlabel('Days with respect to B-band peak')
        ax.legend(fontsize=14)

        plt.show()
    
    return reducer, transformed_features


def plot_coef(components_df, params_df):

    n_components = components_df.shape[1]
    fig = plt.figure(figsize=(12, 12))
    gs = gridspec.GridSpec(n_components , n_components, wspace=0, hspace=0)

    for i in np.arange(n_components):

        ax = plt.subplot(gs[i, 0])
        if i==0:
            ax.set_ylabel(f'# SNe')
        else:
            ax.set_ylabel(f'Coef. {int(i+1)}')
        for j in np.arange(i+1):

            ax = plt.subplot(gs[i, j])
            if i==j:
                ax.hist(components_df.values[:,i])
                if i!=n_components-1:
                    ax.set_xticklabels([' ']*len(components_df.values[:,i]))
            else:    
                s = ax.scatter(components_df.values[:,j], components_df.values[:,i])

            if (i+1)==n_components:
                ax = plt.subplot(gs[i, j])
                ax.set_xlabel(f'Coef. {int(j+1)}')
            if j != 0:
                ax.set_yticklabels([' ']*len(components_df.values[:,i]))

    #plt.savefig(f'plots/{band}-band_coefficients.pdf')
    plt.show()


def plot_coefvparams(components_df, params_df, param_names=None):
    
    if param_names is None:
        param_names = [r'm$_B$', '$\Delta$m$_{15}$', r'($B$ - $V$)$_{\rm max}$', 'M$_B$']
        
    n_components = components_df.shape[1]
    fig, ax = plt.subplots(n_components, params_df.shape[1], figsize=(12, 10), 
                           sharex='col', sharey='row', gridspec_kw={'wspace':0, 'hspace':0})

    for i in np.arange(n_components):

        ax[i, 0].set_ylabel(r'Coef. %i'%(i+1))
    
        for j in np.arange(params_df.shape[1]):

            ax[i, j].scatter(params_df.values[:,j], components_df.values[:,i])

            if (i+1)==n_components:
                ax[i, j].set_xlabel(f'{param_names[j]}')  

    #plt.savefig(f'plots/coef_v_lcparams.pdf')
    plt.show()
    
def cosmo_mcmc(p0, args, labels=None, nwalkers=32, nsteps=1000, nburn=500, verbose=True, plot=False):

    def log_prior(theta):

        M, alpha, beta = theta
        if -20.0 < M < -17 and -5.0 < alpha < 5.0 and -5.0 < beta < 5.0:
            return 0.0
        return -np.inf

    def log_likelihood(theta, mb, x1, color, dmb, dx1, dcolor, z, dz=0, sig_int=0,
                       sig_lensing=0, sig_hostcorr=0, cov_m_s=0, cov_m_c=0, cov_s_c=0):

        M, alpha, beta = theta
        mu_model = mb - M + alpha*x1 - beta*color
        mu = cosmo.distmod(z).value

        sigma2 = (dmb**2 + (alpha**2)*(dx**2) + (beta**2)*(dy**2) 
                         + (5*dz/(z*np.log(10)))**2
                         + sig_lensing**2 + sig_hostcorr + sig_int**2
                         + 2*alpha*cov_m_s + 2*(-beta)*cov_m_c + 2*alpha*(-beta)*cov_s_c
                 )
        return -0.5*np.sum((mu-mu_model)**2/sigma2 + np.log(sigma2))

    def log_probability(theta, mb, x1, color, dmb, dx1, dcolor, z, dz=0, sig_int=0, 
                        sig_lensing=0, sig_hostcorr=0, cov_m_s=0, cov_m_c=0, cov_s_c=0):

        lp = log_prior(theta)
        if not np.isfinite(lp):
            return -np.inf
        return lp + log_likelihood(theta, mb, x1, color, dmb, dx1, dcolor, z, dz=0, sig_int=0,
                                   sig_lensing=0, sig_hostcorr=0, cov_m_s=0, cov_m_c=0, cov_s_c=0)

    ndim = len(p0)
    pos = p0 + 1e-4*np.random.randn(nwalkers, ndim)

    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=args)
    sampler.run_mcmc(pos, nsteps);
    samples = sampler.chain[:, nburn:, :].reshape((-1, ndim))
    
    if plot==True:
        if labels is None:
            labels = [f'param{i}' for i in ndim]
        fig = corner.corner(samples, labels=labels, truths=p0);
        
    results = list()
    for i in range(ndim):
        mcmc = np.percentile(samples[:, i], [16, 50, 84])
        results.append(mcmc[1])
        q = np.diff(mcmc)
        txt = "\mathrm{{{3}}} = {0:.3f}_{{-{1:.3f}}}^{{{2:.3f}}}"
        txt = txt.format(mcmc[1], q[0], q[1], labels[i].replace('$', ''))
        if verbose==True:
            display(Math(txt))
        
    return results
```

### Conley et al. (2011)
\begin{aligned} \mathbf{D}_{\text {stat }, i i}=& \sigma_{m_{B}, i}^{2}+\alpha^{2} \sigma_{s, i}^{2}+\beta^{2} \sigma_{\mathcal{C}, i}^{2}+\sigma_{\text {int }}^{2}+\left(\frac{5\left(1+z_{i}\right)}{z_{i}\left(1+z_{i} / 2\right) \log 10}\right)^{2}\times \sigma_{z, i}^{2}+\sigma_{\text {lensing }}+\sigma_{\text {hostcorrection }}+C_{m_{B} s C, i} \end{aligned}

### Betoule et al. (2014)
\begin{aligned} C_{\eta}=C_{\text {stat }} &+\left(C_{\text {cal }}+C_{\text {model }}+C_{\text {bias }}+C_{\text {host }}+C_{\text {dust }}\right)_{\text {reevaluated }} \\ &+\left(C_{\text {pecvel }}+C_{\text {nonla }}\right)_{C 11} \end{aligned}
$$ \mathrm{C}=\mathrm{AC}_{\eta} \mathrm{A}^{\dagger}+\operatorname{diag}\left(\frac{5 \sigma_{z}}{z \log 10}\right)^{2}+\operatorname{diag}\left(\sigma_{\mathrm{lens}}^{2}\right)+\operatorname{diag}\left(\sigma_{\mathrm{coh}}^{2}\right)$$

```python
def residual(params, mb, x1, color, dmb, dx1, dcolor, mass_step, z, dz=0, sig_int=0, 
             sig_lensing=0, sig_hostcorr=0, cov_m_s=0, cov_m_c=0, cov_s_c=0):
    
    alpha = params['alpha'].value
    beta = params['beta'].value
    M = params['M'].value
    
    mu = cosmo.distmod(z).value
    mu_corr = mb - (M + mass_step) + alpha*x1 - beta*color
    
    D = (dmb**2 + (alpha**2)*(dx1**2) + (beta**2)*(dcolor**2)
         + (5*dz/(z*np.log(10)))**2
         + sig_lensing**2 + sig_hostcorr**2 + sig_int**2
         + 2*alpha*cov_m_s + 2*(-beta)*cov_m_c + 2*alpha*(-beta)*cov_s_c
        )
    
    dof = len(mu) - 3
    chi2 = np.sum((mu - mu_corr)**2/D)
    
    return chi2
```

```python
jla_params = pd.read_csv('test_files/JLA/jla_lcparams.txt', delim_whitespace=True)

z = jla_params.zhel.values
dz = 0.0  # included in dmb
mu = cosmo.distmod(z).value

mb = jla_params.mb.values  # biascor is already applied to mb, original measurements can be obtained by subtracting this term to mb
dmb = jla_params.dmb.values  # includes contributions from intrinsic dispersion, lensing, and redshift uncertainty
x1 = jla_params.x1.values
dx1 = jla_params.dx1.values
color = jla_params.color.values
dcolor = jla_params.dcolor.values
rdvar = jla_params['3rdvar'].values
drdvar = jla_params['d3rdvar'].values

# Mass step = -0.061 +/- 0.012 as in B14
mass_step = np.asarray([-0.061 if log_mass>=10 else 0.0 for mag, log_mass in zip(mb, rdvar)])

cov_m_s = jla_params.cov_m_s.values
cov_m_c = jla_params.cov_m_c.values
cov_s_c = jla_params.cov_s_c.values

alpha = 0.141  # B14
beta = 3.101  # B14
M =-19.05  # B14
mu_corr1 = mb - (M + mass_step) + alpha*x1 - beta*color 

int_dict = {1:0.080, 2:0.108, 3:0.134, 4:0.100}  # B14, SNLS, SDSS, lowz, HST
sig_int = 0  # included in dmb

sig_lensing = 0.0  # included in dmb
sig_hostcorr = 0.012  # mass step uncertainty

D = (dmb**2 + (alpha**2)*(dx1**2) + (beta**2)*(dcolor**2)
            + (5*dz/(z*np.log(10)))**2
            + sig_lensing**2 + sig_hostcorr**2 + sig_int**2
            + 2*alpha*cov_m_s + 2*(-beta)*cov_m_c + 2*alpha*(-beta)*cov_s_c
    )

dof = len(mu) - 5  # 735 (includes alpha, beta, M(H0), OmegaM and Mass step)
chi2 = np.sum((mu - mu_corr1)**2/D)
red_chi2 = np.round(chi2/dof, 2)
sig = np.round(np.sqrt(2/dof), 2)
print('Chi =', red_chi2, '+/-', sig)
print(f'Residual = {np.average(mu - mu_corr1, weights=1/D):.2f} +/- {np.std(mu - mu_corr1):.2f}')

plt.plot(z, mu_corr1-mu, 'ko', alpha=0.3, label='$x_1 + C$ (Betoule et al. 2014)')
plt.plot(z, mu-mu, 'r-')
plt.xlabel('$z$', fontsize=26)
plt.ylabel(r'$\mu$ - $\mu_{\Lambda \rm CDM}$', fontsize=26)
plt.legend(fontsize=14)
plt.show()
```

```python
def residual2(params, m, b1, b2, b3, dm, db1, db2, db3, z, dz=0, sig_int=0, 
             sig_lensing=0, sig_hostcorr=0, cov_m_b1=0, cov_m_b2=0, cov_m_b3=0,
             cov_b1_b2=0, cov_b1_b3=0, cov_b2_b3=0):
    
    alpha = params['alpha'].value
    beta = params['beta'].value
    gamma = params['gamma'].value
    M = params['M'].value
    
    mu = cosmo.distmod(z).value
    mu_corr = m - M + alpha*b1 - beta*b2 - gamma*b3
    
    D = (dm**2 + (alpha**2)*(db1**2) + (beta**2)*(db2**2) + (gamma**2)*(db3**2)
            + (5*dz/(z*np.log(10)))**2
            + sig_lensing**2 + sig_hostcorr**2 + sig_int**2
            + 2*alpha*cov_m_b1 + 2*(-beta)*cov_m_b2 + 2*(-gamma)*cov_m_b3 
            + 2*alpha*(-beta)*cov_b1_b2 + 2*alpha*(-gamma)*cov_b1_b3
            + 2*(-beta)*(-gamma)*cov_b2_b3
           )
    
    dof = len(mu) - 4
    chi2 = np.sum((mu - mu_corr)**2/D)
    
    return chi2

def cosmo_mcmc2(p0, args, labels=None, nwalkers=32, nsteps=1000, nburn=500, verbose=True, plot=False):

    def log_prior(theta):

        M, alpha, beta, gamma = theta
        if -20.0 < M < -17 and -5.0 < alpha < 5.0 and -5.0 < beta < 5.0 and -5.0 < gamma < 5.0:
            return 0.0
        return -np.inf

    def log_likelihood(theta, m, b1, b2, b3, dm, db1, db2, db3, z, dz=0, sig_int=0, 
                     sig_lensing=0, sig_hostcorr=0, cov_m_b1=0, cov_m_b2=0, cov_m_b3=0,
                     cov_b1_b2=0, cov_b1_b3=0, cov_b2_b3=0):

        M, alpha, beta, gamma = theta
        mu_model = m - M + alpha*b1 - beta*b2 - gamma*b3
        mu = cosmo.distmod(z).value

        sigma2 = (dm**2 + (alpha**2)*(db1**2) + (beta**2)*(db2**2) + (gamma**2)*(db3**2)
                        + ( (5*(1+z)/(z*(1+z/2)*np.log(10)))**2 )*(dz**2) 
                        + sig_lensing**2 + sig_hostcorr**2 + sig_int**2
                        + 2*alpha*cov_m_b1 + 2*(-beta)*cov_m_b2 + 2*(-gamma)*cov_m_b3 
                        + 2*alpha*(-beta)*cov_b1_b2 + 2*alpha*(-gamma)*cov_b1_b3
                        + 2*(-beta)*(-gamma)*cov_b2_b3
                       )
        return -0.5*np.sum((mu-mu_model)**2/sigma2 + np.log(sigma2))

    def log_probability(theta, m, b1, b2, b3, dm, db1, db2, db3, z, dz=0, sig_int=0, 
                     sig_lensing=0, sig_hostcorr=0, cov_m_b1=0, cov_m_b2=0, cov_m_b3=0,
                     cov_b1_b2=0, cov_b1_b3=0, cov_b2_b3=0):

        lp = log_prior(theta)
        if not np.isfinite(lp):
            return -np.inf
        return lp + log_likelihood(theta, m, b1, b2, b3, dm, db1, db2, db3, z, dz, sig_int, 
                                 sig_lensing, sig_hostcorr, cov_m_b1, cov_m_b2, cov_m_b3,
                                 cov_b1_b2, cov_b1_b3, cov_b2_b3)

    ndim = len(p0)
    pos = p0 + 1e-4*np.random.randn(nwalkers, ndim)

    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=args)
    sampler.run_mcmc(pos, nsteps);
    samples = sampler.chain[:, nburn:, :].reshape((-1, ndim))
    
    if plot==True:
        if labels is None:
            labels = [f'param{i}' for i in ndim]
        fig = corner.corner(samples, labels=labels, truths=p0);
        
    results = list()
    for i in range(ndim):
        mcmc = np.percentile(samples[:, i], [16, 50, 84])
        results.append(mcmc[1])
        q = np.diff(mcmc)
        txt = "\mathrm{{{3}}} = {0:.3f}_{{-{1:.3f}}}^{{{2:.3f}}}"
        txt = txt.format(mcmc[1], q[0], q[1], labels[i].replace('$', ''))
        if verbose==True:
            display(Math(txt))
        
    return results


components = ['c2', 'c3']
labels = f'${components[0][0]}_{components[0][1]} + {components[1][0]}_{components[1][1]} + C$'
cpx = components_df2[components[0]].values
cpy = components_df2[components[1]].values

biascor = []
for sn in components_df2.index:
    if sn in jla_params['#name'].values:
        biascor.append(jla_params[jla_params['#name']==sn].biascor.values[0])
    else:
        biascor.append(0)   
biascor = np.asarray(biascor)

z0 = dparams_df2.z.values
dz = 0.0005
mu0 = cosmo.distmod(z0).value

mb = params_df2.mb.values + biascor
dmb = dparams_df2.dmb.values

dm15 = params_df2.dm15.values
dm15err = dparams_df2.dm15err.values
dm15_mask = np.where(dm15>-100)

color = params_df2.color.values
dcolor = dparams_df2.dcolor.values
color_mask = np.where(color>-100)

cov_m_b1=cov_m_b2=cov_m_b3=cov_b1_b2=cov_b1_b3=cov_b2_b3=0

X = (cpx - cpx.mean())
Y = (cpy - cpy.mean())
dX = dY = 0.0

### error budget ###
int_dict = {'s':0.080, 'S':0.108, '0':0.134}  # B14, SNLS, SDSS, lowz
sig_int =  np.asarray([int_dict[sample[0]] for sample in params_df2.index.values])

sig_lensing = 0.055*z0
sig_hostcorr = 0.0  # mass step uncertainty
####################

mb = mb[color_mask]
dmb = dmb[color_mask]
mu0 = mu0[color_mask]
z0 = z0[color_mask]

sig_int = sig_int[color_mask]
sig_lensing = sig_lensing[color_mask]

X = (cpx - cpx.mean())[color_mask]
Y = (cpy - cpy.mean())[color_mask]
dX = dY = 0.0
Z = color[color_mask]
dZ = dcolor[color_mask]

##########################################

args=(mb, X, Y, Z, dmb, dX, dY, dZ, z0, dz, sig_int, 
      sig_lensing, sig_hostcorr, cov_m_b1, cov_m_b2, cov_m_b3,
      cov_b1_b2, cov_b1_b3, cov_b2_b3)
# methods: leastsq, least_squares, nelder, bfgs
#result = lmfit.minimizer.minimize(fcn=residual2, params=params, args=args, method='nelder')

minim = 'lmfit'

if minim=='lmfit':
    params = lmfit.Parameters()
    params.add('alpha', value=0.0, min=-5, max=5)  
    params.add('beta', value=1, min=-5, max=5)
    params.add('gamma', value=1, min=-5, max=5)
    params.add('M', value=-19, min=-21, max=-17)
    # methods: leastsq, least_squares, nelder, bfgs
    result = lmfit.minimizer.minimize(fcn=residual2, params=params, args=args, method='nelder')
    M, alpha = result.params['M'].value, result.params['alpha'].value
    beta, gamma = result.params['beta'].value, result.params['gamma'].value
    
elif minim=='mcmc':
    p0 = np.asarray([-19, 1, 2, 1])
    labels = ['M', r'$\alpha$', r'$\beta$', r'$\gamma$']
    results = cosmo_mcmc2(p0, args, labels=labels, nsteps=500, nburn=200, verbose=False)
    M, alpha, beta, gamma = results

print(f'M = {M:.2f}, alpha = {alpha:.2f}, beta = {beta:.2f}, gamma = {gamma:.2f}') 
mu_corr0 = mb - M + alpha*X - beta*Y - gamma*Z
    
D = (dmb**2 + (alpha**2)*(dX**2) + (beta**2)*(dY**2) + (gamma**2)*(dZ**2)
            + (5*dz/(z0*np.log(10)))**2
            + sig_lensing**2 + sig_hostcorr**2 + sig_int**2
            + 2*alpha*cov_m_b1 + 2*(-beta)*cov_m_b2 + 2*(-gamma)*cov_m_b3 
            + 2*alpha*(-beta)*cov_b1_b2 + 2*alpha*(-gamma)*cov_b1_b3
            + 2*(-beta)*(-gamma)*cov_b2_b3
           )

dof = len(mu0) - 4
chi2 = np.sum((mu0 - mu_corr0)**2/D)
red_chi2 = np.round(chi2/dof, 2)
sig = np.round(np.sqrt(2/dof), 2)
print('Chi =', red_chi2, '+/-', sig)
print(f'Residual = {np.average(mu0 - mu_corr0, weights=1/D):.2f} +/- {np.std(mu0 - mu_corr0):.2f}')
print(f'{len(mu0)} SNe')

plt.plot(z0, mu_corr0 - mu0, 'ko', alpha=0.3, label=labels)
plt.plot(z0, mu0-mu0, 'r-')
plt.xlabel('$z$', fontsize=26)
plt.ylabel(r'$\mu$ - $\mu_{\Lambda \rm CDM}$', fontsize=26)
plt.legend(fontsize=14)
plt.show()
```

```python
components = ['c1', 'c2']
cpx = components_df[components[0]].values
cpy = components_df[components[1]].values

biascor = []
for sn in components_df.index:
    if sn in jla_params['#name'].values:
        biascor.append(jla_params[jla_params['#name']==sn].biascor.values[0])
    else:
        biascor.append(0)   
biascor = np.asarray(biascor)

z0 = dparams_df.z.values
dz = 0.0001
mu0 = cosmo.distmod(z0).value

mb = params_df.mb.values + biascor
dmb = dparams_df.dmb.values

dm15 = params_df.dm15.values
dm15err = dparams_df.dm15err.values
dm15_mask = np.where(dm15>-100)

color = params_df.color.values
dcolor = dparams_df.dcolor.values
color_mask = np.where(color>-100)

cov_m_s = 0
cov_m_c = 0
cov_s_c = 0

x = (cpx - cpx.mean())
y = (cpy - cpy.mean())
dx = dy = 0.0

### error budget ###
int_dict = {'s':0.080, 'S':0.108, '0':0.134}  # B14, lowz, SDSS, SNLS
sig_int =  np.asarray([int_dict[sn_name[0]] for sn_name in params_df.index.values])

sig_lensing = 0.055*z0
sig_hostcorr = 0.0
####################

params = lmfit.Parameters()
params.add('alpha', value=0.0, min=-5, max=5)  
params.add('beta', value=1, min=-5, max=5)  
params.add('M', value=-19, min=-21, max=-17)

args=(mb, x, y, dmb, dy, dx, z0, dz, sig_int, 
      sig_lensing, sig_hostcorr, cov_m_s, cov_m_c, cov_s_c)

minim = 'lmfit'

if minim=='lmfit':
    params = lmfit.Parameters()
    params.add('alpha', value=0.0, min=-5, max=5)  
    params.add('beta', value=1, min=-5, max=5)  
    params.add('M', value=-19, min=-21, max=-17)
    # methods: leastsq, least_squares, nelder, bfgs
    result = lmfit.minimizer.minimize(fcn=residual, params=params, args=args, method='nelder')
    M, alpha, beta = result.params['M'].value, result.params['alpha'].value, result.params['beta'].value
    
elif minim=='mcmc':
    p0 = np.asarray([-19, 1, 2])
    labels = ['M', r'$\alpha$', r'$\beta$']
    results = cosmo_mcmc(p0, args, labels=labels, nsteps=500, nburn=200, verbose=False)
    M, alpha, beta = results

print(f'M = {M:.2f}, alpha = {alpha:.2f}, beta = {beta:.2f}')    
mu_corr0 = mb - M + alpha*x - beta*y
    
D = (dmb**2 + (alpha**2)*(dx**2) + (beta**2)*(dy**2) 
            + ( (5*(1+z0)/(z0*(1+z0/2)*np.log(10)))**2 )*(dz**2) 
            + sig_lensing**2 + sig_hostcorr + sig_int**2
            + 2*alpha*cov_m_s + 2*(-beta)*cov_m_c + 2*alpha*(-beta)*cov_s_c
    )

dof = len(mu0) - 3
chi2 = np.sum((mu0 - mu_corr0)**2/D)
red_chi2 = np.round(chi2/dof, 2)
sig = np.round(np.sqrt(2/dof), 2)
print('Chi =', red_chi2, '+/-', sig, f', std = {np.std(mu0 - mu_corr0):.2f}')

plt.plot(z0, mu0-mu_corr0, 'ko', alpha=0.3, label=f'{components[0]} + {components[1]} (This work)')
plt.plot(z0, mu0-mu0, 'r-')
plt.legend(fontsize=14)
plt.show()
```
